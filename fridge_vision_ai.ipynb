```python
import os
import base64
from mimetypes import guess_type
from openai import OpenAI

# API key
client = OpenAI(api_key=os.getenv("not_Real_Key"))

# Helper to encode a local image file into base64 data URL
def encode_image_to_data_url(image_path: str) -> str:
    mime_type, _ = guess_type(image_path)
    mime_type = mime_type or "application/octet-stream"
    with open(image_path, "rb") as f:
        b64 = base64.b64encode(f.read()).decode("utf-8")
    return f"data:{mime_type};base64,{b64}"

# Gave Variable a path file to the image 
data_url = encode_image_to_data_url("First Project/fridge.png")

# GPT‑4o prompt (what I want to say to it)
messages = [
    {"role": "system", "content": "You are an assistant that can identify foods from an image and suggest recipes."},
    {"role": "user", "content": [
        {"type": "text", "text": "Please identify the foods and approximate quantities in this image, then give simple recipes you can make with them."},
        {"type": "image_url", "image_url": {"url": data_url}}
    ]}
]

# Calling GPT‑4o model
response = client.chat.completions.create(
    model="gpt-4o",
    messages=messages,
    max_tokens=400,
    temperature=0.7
)

# output
print(response.choices[0].message.content)

```
